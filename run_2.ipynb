```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read CSV files
csv_path_1 = 'C:\\Users\\andre\\development\\pole_detection_2\\runs\\detect\\train11\\results.csv'
csv_path_2 = 'C:\\Users\\andre\\development\\pole_detection_1\\runs\\detect\\train10\\results.csv'
csv_path_3 = 'C:\\Users\\andre\\Downloads\\Pole Detection\\runs\\detect\\train6\\results.csv'
# csv_path_4 = 'C:\\Users\\andre\\development\\pole_detection_3\\runs\\detect\\train8\\results.csv'  # Changed variable name here

from PIL import Image

f1_curve_path = 'C:\\Users\\andre\\development\\pole_detection_2\\runs\\detect\\train11\\F1_curve.png'
f1_curve_img = Image.open(f1_curve_path)
f1_curve_img = f1_curve_img.resize((300, 300)) 

def read_and_clean_csv(csv_path):
    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip()  # Strip the whitespace from column names
    return df

df1 = read_and_clean_csv(csv_path_1)
df2 = read_and_clean_csv(csv_path_2)
df3 = read_and_clean_csv(csv_path_3)
# df4 = read_and_clean_csv(csv_path_4)

# Combine dataframes
df1['model'] = 'Model 1'
df2['model'] = 'Model 2'
df3['model'] = 'Model 3'
# df4['model'] = 'Model 4'

combined_df = pd.concat([df1, df2, df3], axis=0, ignore_index=True)

# Function to plot comparison
def plot_comparison(df, metric, title, ax):
    for model in df['model'].unique():
        model_df = df[df['model'] == model]
        ax.plot(model_df['epoch'], model_df[metric], label=model)
    ax.set_xlabel('Epoch')
    ax.set_ylabel(metric)
    ax.set_title(title)
    ax.legend()

# Create a 4x3 grid of subplots
fig, axes = plt.subplots(4, 3, figsize=(15, 20))

# Loss values comparison
plot_comparison(combined_df, 'train/box_loss', 'Box Loss Comparison', axes[0, 0])
plot_comparison(combined_df, 'train/cls_loss', 'Classification Loss Comparison', axes[0, 1])
plot_comparison(combined_df, 'train/dfl_loss', 'DFL Loss Comparison', axes[0, 2])

# mAP scores comparison
plot_comparison(combined_df, 'metrics/mAP50(B)', 'mAP50 Comparison', axes[1, 0])
plot_comparison(combined_df, 'metrics/mAP50-95(B)', 'mAP50-95 Comparison', axes[1, 1])

# Learning rate comparison
plot_comparison(combined_df, 'lr/pg0', 'Learning Rate pg0 Comparison', axes[1, 2])
plot_comparison(combined_df, 'lr/pg1', 'Learning Rate pg1 Comparison', axes[2, 0])
plot_comparison(combined_df, 'lr/pg2', 'Learning Rate pg2 Comparison', axes[2, 1])

# Precision and recall comparison
plot_comparison(combined_df, 'metrics/precision(B)', 'Precision Comparison', axes[2, 2])
plot_comparison(combined_df, 'metrics/recall(B)', 'Recall Comparison', axes[3, 0])

# Bar chart visualization with explanations
# X values and bar width
x = np.arange(len(combined_df['model'].unique()))
width = 0.25

# Compute the average metric values for each model
model_metrics = combined_df.groupby('model').mean()

# Create a bar chart on the last subplot
ax = axes[3, 1]
rects1 = ax.bar(x - width, model_metrics['metrics/mAP50(B)'], width, label='mAP50')
rects2 = ax.bar(x, model_metrics['metrics/mAP50-95(B)'], width, label='mAP50-95')
rects3 = ax.bar(x + width, model_metrics['metrics/recall(B)'], width, label='Recall')

ax.set_ylabel('Scores')
ax.set_title('Model Comparison')
ax.set_xticks(x)
ax.set_xticklabels(combined_df['model'].unique())
ax.legend()

# Annotate the bars with their corresponding values
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# Make space for the bar chart subplot and display the figure
fig.tight_layout()
plt.show()
```

```python
# Calculate F1 scores
combined_df['f1_score'] = 2 * (combined_df['metrics/precision(B)'] * combined_df['metrics/recall(B)']) / (combined_df['metrics/precision(B)'] + combined_df['metrics/recall(B)'])

# Compute the average F1 score for each model
model_f1_scores = combined_df.groupby('model')['f1_score'].mean()
```

### The F1-score is a measure of a model's accuracy, calculated by using the following formula: 
F1 = 2*(Precision*Recall)/(Precision+Recall). 

It provides a balance between precision and recall, and a weighted average can be used to evaluate the overall performance of a model.

```python
# Create a 3x1 grid of subplots with the aspect ratio of the images
fig, axes = plt.subplots(3, 1, figsize=(5 * aspect_ratio, 15))

# Display F1 curves in subplots and add F1 scores
axes[0].imshow(f1_curve_img_1)
axes[0].set_title('Model 1 F1 Curve')
axes[0].text(1.05, 0.5, f'F1-score: {model_f1_scores["Model 1"]:.2f}', fontsize=22, color='blue', transform=axes[0].transAxes)
axes[0].axis('off')

axes[1].imshow(f1_curve_img_2)
axes[1].set_title('Model 2 F1 Curve')
axes[1].text(1.05, 0.5, f'F1-score: {model_f1_scores["Model 2"]:.2f}', fontsize=22, color='blue', transform=axes[1].transAxes)
axes[1].axis('off')

axes[2].imshow(f1_curve_img_3)
axes[2].set_title('Model 3 F1 Curve')
axes[2].text(1.05, 0.5, f'F1-score: {model_f1_scores["Model 3"]:.2f}', fontsize=22, color='blue', transform=axes[2].transAxes)
axes[2].axis('off')

# Show the plot
plt.show()
```
